```python
df = pd.read_csv("C:/Users/82108/Desktop/MyProject/백준추천시스템/data/user_data_with_problem_for_EASE.csv")
over6 = df[df["tier"] >= 6]
down15 = df[df["tier"] <= 15]
silver_gold = pd.merge(over6,down15,how="inner")
silver_gold_half = silver_gold[silver_gold["Unnamed: 0"] % 2 ==0]
```

```python
silver_gold_half.reset_index(inplace=True,drop=True)
```

```python
idx_to_id = {}
for i in range(len(silver_gold_half)):
    idx_to_id[i] = silver_gold_half.loc[i,"handle"]
```

```python
 df_result.to_csv("중간결과_todo_IDXtoID.csv")
```

```python
df_result.rename(columns={1:"recommended_rival"},inplace=True)
df_result.rename(columns={0:"user_id"},inplace=True)
```

```python
def conver_idx_to_id(id_list):
    for i in range(len(id_list)):
        id_list = list(map(str,id_list))
        id_list[i] = idx_to_id[int(id_list[i])]
    return id_list
```

```python
df_result['recommended_rival'] = df_result['recommended_rival'].apply(conver_idx_to_id)
```

```python
for i in range(len(df_result)):
    df_result.iloc[i,0] = idx_to_id[i]
```
